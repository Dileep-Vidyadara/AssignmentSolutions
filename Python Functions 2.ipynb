1.Reverse a given list in Python <br>
    `list1 = [100, 200, 300, 400, 500]` <br>
    `Expected output: [500, 400, 300, 200, 100]`

```
list1 = [100, 200, 300, 400, 500]
list1.reverse()
print (list1)
```

2.Concatenate two lists index-wise <br>
     list1 = ["M", "na", "i", "Ke"] <br>
     list2 = ["y", "me", "s", "lly"] <br>
     Expected output: ['My', 'name', 'is', 'Kelly']

```
list1 = ["M", "na", "i", "Ke"]
list2 = ["y", "me", "s", "lly"] 

list_concate = [i + j for i, j in zip(list1, list2)]

print(list_concate)
```

3.Given a two Python list. Iterate both lists simultaneously such that list1 should display item in original order and list2 in reverse order <br>
    `list1 = [10, 20, 30, 40]` <br>
    `list2 = [100, 200, 300, 400]` <br>
   `Expected output:` <br> 
           `10 400` <br>
`20 300` <br>
`30 200` <br>
`40 100` <br>

```
list1 = [10, 20, 30, 40]
list2 = [100, 200, 300, 400]
 
for (i,j) in zip( list1, reversed(list2) ):
    print(i,j)
```

4.Remove empty strings from the list of strings <br>
          list1 = ["Mike", "", "Emma", "Kelly", "", "Brad"]

```
list1 = ["Mike", "", "Emma", "Kelly", "", "Brad"]

while("" in list1) :
    list1.remove("")
    
print(list1)
```

5.Below are the two lists convert it into the dictionary <br>
        `keys = ['Ten', 'Twenty', 'Thirty']` <br>
        `values = [10, 20, 30]` <br>
        `Expected output: {'Ten': 10, 'Twenty': 20, 'Thirty': 30}` <br>

```
dkeys = ['Ten', 'Twenty', 'Thirty']
dvalues = [10, 20, 30]

dictry = {}
for i in dkeys:
    for j in dvalues:
        dictry[i] = j

print(dictry)
```

6.Merge following two Python dictionaries into one <br>
   dict1 = {'Ten': 10, 'Twenty': 20, 'Thirty': 30} <br>
   dict2 = {'Thirty': 30, 'Fourty': 40, 'Fifty': 50} <br>
   Expected output: {'Ten': 10, 'Twenty': 20, 'Thirty': 30, 'Fourty': 40, 'Fifty': 50}

```
dict1 = {'Ten': 10, 'Twenty': 20, 'Thirty': 30} 
dict2 = {'Thirty': 30, 'Fourty': 40, 'Fifty': 50} 

dict_merge = {**dict1, **dict2}

print(dict_merge)
```

8.Parse the following JSON to get all the values of a key ‘name’ within an array <br>
` [ {"id" : 1, "name" : "name1", "color" : [ "red", "green" ] }, { "id" : 2, "name" : "name2", "color" : ["pink","yellow"] } ] ` <br>
` Expected output: ["name1", "name2"] `

```
data = [ {"id" : 1, "name" : "name1", "color" : [ "red", "green" ] }, 
        { "id" : 2, "name" : "name2", "color" : ["pink","yellow"] } ]

data_new = []
for i in range(len(data)):
    data_new.append(data[i]['name'])
    
print(data_new)
```
